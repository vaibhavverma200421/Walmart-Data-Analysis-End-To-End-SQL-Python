Data Exploration And loading (cleaning, Transforming, Removing duplicates and null values etc)



```python
# Importing Dependencies
 
import pandas as pd 

# PostgreSQL connection


import psycopg2    # it will help to connect python with postgreSQL database and work as a adapter
from sqlalchemy import create_engine

```

```python
df = pd.read_csv("Walmart.csv", encoding_errors='ignore')

df.shape
```

```python
# let's check the first 5 rows of our dataset

df.head(5)
```

```python
# getting a little more information about our dataset.

df.describe()
#this feature helps us to get description to get the aggregations like count,mean,standard deviation, min, max count etc.

df.info()
# this info() help us to identify null columns if exist, datatypes and memory usage 

# converting all column names to lowercase

df.columns = df.columns.str.lower()

# ChecK 
df.head(3)
```

```python
df.isnull().sum()

# as we can see there are two columns unit_price and quantity that have 31 nulls each as shown in the output below 

# we will deal with this after removing the duplicates from the dataset
```

```python
# now we are going to check foe duplicated records for this we use a function 

df.duplicated().sum()

# as the output shows we have 51 duplicates 
```

```python
# now as we got  the duplicates in the previous cell we are going to remove those duplicates.

df.drop_duplicates(inplace= True)  # all the duplicates are  removed from the dataframe when i execute it.

# now let's verify or see if the duplicates are removed from the dataframe
df.duplicated().sum()

# so as you can see the output is 0 which means that all duplicates are removed from the dataframe 
```

```python
# now we are going to drop all the null values from the dataset
# for this we use :

df.dropna(inplace= True)   # all the null values are removed from the dataset


# let's verify if the null values are removed from the dataset
df.isnull().sum()
```

```python
# now let's also see the data types of different columns to see if they are correct 
df.dtypes

# as we can see the unit_price column is in object datatype we need to convert it into 
# float datatype for this we will remove the $ sign from the column and then convert it into float datatype

df['unit_price'] = pd.to_numeric(df['unit_price'].astype(str).str.replace('$', '', regex=False),
errors='coerce')






# # let's verify if the changes are done
# df.dtypes

# # so now we can clearly see that the unit_price column is now in float datatype.
# df['unit_price'].head(5)

```

```python

#  Creating a new column "Total_Price"

df['Total_Price'] = df['unit_price'] * df['quantity']   

df.head(5)

```

The cleaning and transformation is done.Let's export the data to SQL now.

```python
# # exporting the cleaned data to a new csv file
df.to_csv("walmart_cleaned_data.csv", index=False)
```
